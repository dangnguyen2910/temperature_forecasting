```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/paris_temperature.csv')
df.head()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df = df.set_index(df['datetime'])
```

```python
fig = plt.figure(figsize=(15,5))
fig.add_subplot(2,1,1)
plt.plot(df['temp'])
plt.title("Paris temperature per day")
```

```python
# Loop through each column and print unique values
def uni_val(df, col):
    unique_values = df[col].unique()
    return unique_values
```

```python
print(uni_val(df, 'preciptype'))
print(uni_val(df, 'conditions'))
print(uni_val(df, 'conditions'))
print(uni_val(df, 'description'))
print(uni_val(df, 'icon'))
print(uni_val(df, 'stations'))
```
